```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib notebook

sinasc = pd.read_csv("sinasc_RO_2019.csv")
sinasc.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(x=sinasc['IDADEMAE'])
plt.title("Boxplot da Idade da Mae")
plt.xlabel("Idade da Mae")
```

####  2) Faça um boxplot do peso do bebê

```python
plt.close('all')
sns.boxplot(sinasc['PESO'])
plt.title("Boxplot Peso do Bebê")
plt.xlabel("Peso do Bebê")
```

####  3) Faça um histograma da idade da mãe

```python
plt.close('all')
sns.displot(data=sinasc['IDADEMAE'], kde=True, bins=20, element='step', alpha = .25)
```

#### 4) Faça um histograma do peso do bebê

```python
plt.close('all')
sns.displot(data=sinasc['PESO'], kde=True, bins=20, element='step', alpha=.25)
```
